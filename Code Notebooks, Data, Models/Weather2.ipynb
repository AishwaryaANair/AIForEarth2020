```python

count = 0
current_key_count = 0
key_track = 0

frequency = 24
final_data = []

previous_n_days_considered = 4
current_key = world_weather_keys[key_track]
max_key_count = 500

requests_time_out = 0
for i in df_values[0:]:
    lat_long = [ str(i[0]) + "," + str(i[1]) ]
    end_date = datetime.strptime(i[2], '%Y-%m-%d').date()
    start_date = end_date - timedelta(previous_n_days_considered)
    one_row = []

    if current_key_count > max_key_count:
        key_track += 1
        current_key = world_weather_keys[key_track]
        current_key_count = 0
    try:
        hist_weather_data = retrieve_hist_data(current_key, lat_long, start_date, end_date, frequency, location_label = False, export_csv = False, store_df = True)
        current_key_count += 1

        for record in hist_weather_data:
            date_times = record["date_time"].values
            maxtempC = record["maxtempC"].values
            sunHour = record["sunHour"].values
            DewPointC = record["DewPointC"].values
            HeatIndexC = record["HeatIndexC"].values
            cloudcover = record["cloudcover"].values
            humidity = record["humidity"].values
            precipMM = record["precipMM"].values
            windspeedKmph = record["windspeedKmph"].values
    
        for each_record in range(len(date_times)):
            current_date = date_times[each_record]
            max_temp = maxtempC[each_record]
            sun_hour = sunHour[each_record]
            dew_point = DewPointC[each_record]
            heat_index = HeatIndexC[each_record]
            cloud_cover = cloudcover[each_record]
            humid = humidity[each_record]
            precp = precipMM[each_record]
            wind_speed = windspeedKmph[each_record]
            fire_or_not = 1 if str(current_date)[0:10] == str(end_date) else 0

            one_row.append([ i[0],i[1], current_date, max_temp, sun_hour, dew_point, heat_index, cloud_cover, humid, precp, wind_speed, fire_or_not])
            
    except HTTPError:
        requests_time_out += 1

    final_data.extend(one_row)  

    count += 1

    if count % 1000 == 0:
        print("Count:{}, Len(Final data):{}, key track:{}".format(count, len(final_data), key_track))

```
